```python
from sympy import Symbol,lambdify

x=Symbol("x")
# function=(x+3)**2
# gradient = lambdify(x,function.diff(x))
# gradient(3)
```

```python
x_value=[]
y_value=[]

def gradient_descent(function, start, learning_rate=0.01, n_iterations=1000):
    #FOD
    gradient = lambdify(x,function.diff(x))
    function = lambdify(x,function)
    
    # when to stop
    precision = 0.000001
        
    i=0
    while(i < n_iterations):
        prev=start
        start = start - (learning_rate * gradient(start))
        
        step_size = abs(prev-start)
        if step_size < precision:
            break
            
        x_value.append(start)
        y_value.append(function(start))
        i=i+1
        print("Iteration: ",i,"\nX value is",start)
        
    print("Local Minima: ",start)
```

```python
function=(x+5)**2
gradient_descent(function,3)
```

```python
import matplotlib.pyplot as plt 

plt.plot(x_value,y_value,'-o')

plt.show()
```
